## Importing 'DDW-C19-0000.XLSX'(Language Data)  and 'DDW-0000C-08.xlsx'(Census Data according to age) file using pandas

```python
import pandas as pd
data = pd.read_excel('INPUT/DDW-C19-0000.XLSX')
census_data = pd.read_excel('INPUT/DDW-0000C-08.xlsx')
```

### Getting the unique list of states/UTs from the census data to check whether they are similar to the ones obtained from the language data

```python
name_lst = list(census_data['Unnamed: 3'][6:].unique())
name_lst1 = name_lst[:]
for i in range(len(name_lst)):
    if name_lst[i] != "INDIA":
        name_lst[i] = name_lst[i].split("- ")[1]
```

### Checking if both have the same states with same sequence

```python
state_lang = list(data['Unnamed: 2'][5:869].unique())
yes = 0
for i in range(len(state_lang)):
    if name_lst[i] == state_lang[i]:
        yes = yes+1
print(yes)
```

### Getting the names of the literacy groups from the language data according to literacy

```python
lit_grp = list(data['Unnamed: 4'][5:869].unique())
lit_grp = lit_grp[1:]
lit_grp.remove('Literate')
lit_grp
```

### Getting the unique names of literacy groups from the census data
### We only need groups from 'Below' which indicates 'Literate but below primary' to 'Graduate & above'

```python
lit_grp_cen = list(census_data.iloc[1].unique())
lit_grp_cen = lit_grp_cen[5:13]
lit_grp_cen
```

### Getting the column names of the columns consisting the values for the literacy groups found above

```python
lit_grp_cen_ind = []
for i in range(len(lit_grp_cen)):
    lit_grp_cen_ind.append(list(census_data.columns[
    (census_data == lit_grp_cen[i])
    .any(axis=0)
    ])[0])
```

### Adding column name of 'Illiterate' to the list

```python
lit_grp_cen_ind.insert(0,'Unnamed: 9')
```

```python
lit_grp_cen_ind
```

```python
lit_grp_tot_cen = [[0]*len(lit_grp)]*len(name_lst1)
for i in range(len(name_lst)):
    temp = [0] * len(lit_grp)
    for j in range(len(lit_grp)):
        if j in range(0,4):
            temp[j] = census_data[lit_grp_cen_ind[j]][(list(census_data['Unnamed: 3']).index(name_lst1[i]))]
        elif j == 4:
            temp[j] = (census_data[lit_grp_cen_ind[j]][(list(census_data['Unnamed: 3']).index(name_lst1[i]))] + 
            census_data[lit_grp_cen_ind[j+1]][(list(census_data['Unnamed: 3']).index(name_lst1[i]))] + 
            census_data[lit_grp_cen_ind[j+2]][(list(census_data['Unnamed: 3']).index(name_lst1[i]))] + 
            census_data[lit_grp_cen_ind[j+3]][(list(census_data['Unnamed: 3']).index(name_lst1[i]))])
        else:
            temp[j] = census_data[lit_grp_cen_ind[j+3]][(list(census_data['Unnamed: 3']).index(name_lst1[i]))]
    lit_grp_tot_cen[i] = temp
```

```python
lit_grp_tot_cen
```

```python
lit_grp_3lang_percent = [[0]*len(lit_grp)]*len(state_lang)
for i in range(len(state_lang)):
    temp = [0]*len(lit_grp)
    for j in range(len(lit_grp)):
        if lit_grp[j] == 'Illiterate':
            temp[j] = (data['Unnamed: 8'][(list(data['Unnamed: 2']).index(state_lang[i]))+j+1])*100 / lit_grp_tot_cen[i][j]
        else:
            temp[j] = (data['Unnamed: 8'][(list(data['Unnamed: 2']).index(state_lang[i]))+j+2])*100 / lit_grp_tot_cen[i][j]
    lit_grp_3lang_percent[i] = temp
```

```python
lit_grp_3lang_percent
```

### Getting the maximum percentage for each state/UT with the index for the age group which has this highest percentage

```python
percent_max = [0] * len(state_lang)
percent_max_litgrp = [0] * len(state_lang)
for i in range(len(name_lst)):
    percent_max[i] = max(lit_grp_3lang_percent[i])
    percent_max_litgrp[i] = lit_grp[lit_grp_3lang_percent[i].index(percent_max[i])]
```

### Getting the codes of the states

```python
state_code = [0] * len(name_lst)
for i in range(len(name_lst)):
    state_code[i] = data['C-19 POPULATION BY BILINGUALISM, TRILINGUALISM, EDUCATIONAL LEVEL AND SEX'][list(data['Unnamed: 2']).index(name_lst[i])]
```

### Exporting the state codes, state names into a dataframe

```python
data_final = {'state-code': state_code, 'state/ut': state_lang,'literacy-group': percent_max_litgrp,'percentage': percent_max}
literacy_india = pd.DataFrame(data_final)
literacy_india
```

```python
literacy_india.to_csv('OUTPUT/literacy-india.csv',index=False)
```
